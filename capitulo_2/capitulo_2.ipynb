## <b>O que é estatística?</b>

### Para motivar nossos estudos de aprendizado estatístico, vamos começar com um exemplo simples. Suponha que somos consultores estatísticos contratados por um determinado cliente para investigar a associação entre anúncios e vendas de um produto em particular. Os dados de anúncio consistem das vendas do produto em 200 mercados diferentes juntamente com a quantidade de dinheiro gastado em anúncio em três mídias diferentes: TV, rádio e jornal. Não é possível para nosso cliente aumentar diretamente a quantidade de vendas do produto. Por outro lado, ele pode controlar o quanto ele gasta em marketing em cada uma das mídias. Portanto, se determinarmos que existe uma associaçào entre marketing e vendas, então podemos intruir nosso cliente para ajustar a quantidade gasta em marketing, indiretamente aumentando as vendas. Em outras palavras, o nosso objetivo é desenvolver um modelo com grande acurácia que pode ser utilizado para prever vendas com base no gasto em cada uma das três mídias.

### Nesse sentido, o valor gasto em marketing são variáveis de input, enquanto "vendas" é uma variável de output. As variáveis de input são tipicamente detnotadas utilizando o símbolo X, com um subscrito para distingui-las. Portanto, X<sub>1</sub> representa o gasto com TV, X<sub>2</sub> o gasto com rádio e X<sub>3</sub> o gasto com jornal. Os inputs recebem diferentes nomes, como preditores, variáveis independentes, features (características), ou as vezes somente variáveis.


### De maneira geral, assumimos que existe uma relação entre Y e X = (X<sub>1</sub>, X<sub>2</sub>, X<sub>3</sub>), o que pode ser escrito na seguinte forma geral:
### Y = f(X) + e

### Aqui f é uma função fixa e desconhecida de X<sub>1</sub>, ..., X<sub>n</sub>, onde e é o termo de erro, que é independente de X e tem média 0. Nessa formulação, f representa a informação sistemática que X provém de Y.

### Esse exemplo mostra um gráfico de salário versus anos de educação. O plot sugere que podemos prever o salário utilizando anos de educação. Porém, essa função f que conecta a variável de input com a variável de output é desconhecida em geral. Já que Income é um dataset simulado, f é conhecido e representado pela linha azul no segundo gráfico. As linhas pretas entre a linha azul e os pontos vermelhos representam o erro de cada um dos dados reais que, em média, somam 0. Em geral, a função f é representada por mais de uma variável de input. Aqui f é uma superfície bidimensional que pode ser estimada com base nos dados observados. Em essência, aprendizado estatístico se refere a um conjunto de approaches para estimar f.

## <b>Por quê estimar f?</b>

### Existem duas principais razões para estimar f: previsão e inferência.

### <b>Previsão</b>

### Em muitas situações, um conjunto de inputs X está dispomível, mas Y não pode ser facilmente obtido. Nesse contexto, já que o a média do erro tende a ser nula, podemos prever Y utilizando
### $$ \hat{Y} = \hat{f}(X)$$

### onde $\hat{f}$ representa nossa estimativa pra f, e $\hat{Y}$ representa a previsão resultante para Y. Nesse contexto, $\hat{f}$ é geralmente tratado como uma caixa preta, no sentido que não estamos necessariamente preocupados com a forma exata de $\hat{f}$, desde que ela retorna previsões suficientemente precisas para Y.

### Como um exemplo, imagine que o conjunto de features X = X<sub>1</sub>, ..., X<sub>p</sub> são as características de um paciente num hospital que podem ser facilmente medidas e Y é uma variável que determina o risco do paciente ter uma reação severe a um determinado tipo de droga. É natural buscar prever Y, tal que se conseguirmos isso seria possível não receitar ao paciente uma droga da qual possivelmente ele terá aversão.

### A acurácia de $\hat{Y}$ como uma previsão para Y pode depender de dois tipos de erros: erros redutíveis e erros irredutíveis. Erros redutíveis são aqueles que podem melhorar a acurácia de $\hat{f}$ em relação à f, tal que podemos reduzir essa taxa de diferença aplicando técnicas estatísticas mais apropriadas. Porém, mesmo que nossa função aproximada $\hat{f}$ seja igual à f, ainda assim existe erro irredutível, que é causado pelo formato da nossa função ($ \hat{Y} = \hat{f}(X) + e$), visto que, por definição, existe um e que não pode ser previsto por X que é, portanto, denominado de erro irredutível. 

```python
import pandas as pd

df = pd.read_csv("datasets/Income2.csv") # esse é o dataset utilizado nessa visualização gráfica
df.head()
```

### Por que o erro irredutível é maior que 0?

```python
import matplotlib.pyplot as plt
import numpy as np


# Create data
education = df["Education"]
seniority = df["Seniority"]
income = df["Income"]

# Create figure and axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.title("Salário por educação e senioridade", fontsize=20)

# Plot data
ax.scatter(education, seniority, income, c='blue')

# Set axis labels
ax.set_xlabel('Anos de educação', fontsize=14)
ax.set_ylabel('Senioridade', fontsize=14)
ax.set_zlabel('Salário',fontsize=14)

plt.subplots_adjust(top = 2, bottom = 1, right = 5, left = -4)
# Show plot
plt.show()
```
