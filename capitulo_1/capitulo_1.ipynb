## <b>Um overview de Estatística</b>

### Estatística se refere a um vasto conjunto de ferramentas para compreender dados. Essas ferramentas podem ser classificadas como supervisionadas e não supervisionadas. Ferramentas supervisionadas de estatística envolvem construir um modelo estatístico para prever, ou estimar, um output baseado em um ou mais inputs. Com estatística não supervisionada, existem inputs mas não existe output supervisionado. Entretanto, podemos entender relações e estruturas de tais dados.

### <b>Dados Salariais</b>

### Examinaremos um número de fatores que interferem no salário de um grupo de homens na região atlântica dos Estados Unidos. Em particular, queremos estudar a relação entre a idade, o nível de educação e o ano do calendário de um dos homens e seu salário. No primeiro gráfico, é possível observar que o salário tende a aumentar até os 60 anos, e depois tende a cair. É mais fácil de observar esse comportamento visualizando a linha azul que representa a média salarial por idade. Porém, é também observável que existe uma grande dispersão no primeiro gráfico, o que indica que idade não é o único fator relevante para prever um salário.

### Os gráficos estão apresentados em salário (milhares de dólares por ano) e suas respectivas grandezas. Pode-se observar no segundo gráfico um aumento de cerca de 10 mil dólares na média salarial após seis anos. Além disso, é claro que a média salarial dos maiores níveis de educação é maior.

### Levando esses dados em consideração, conclui-se que idade, ano analisado e nível educacional são grandezas relevantes para prever um salário.

### <b>Dados do Mercado Financeiro</b>

### A questão do salário envolve prever um valor de output conínuo ou quantitativo. Isso é comumente referenciado como um <i>problema de regressão</i>. Porém, em certos casos, podemos querer prever um valor não numérico; isto é, um output categórico ou qualitativo. Aqui está um exemplo que será aprofundado no capítulo 4:

### No caso anterior, a previsão estatística não envolve prever um valor númerico. Ao invés disso, envolve prever se a performance de um determinado dia no mercado de ações vai ser categorizada no bucket de "subiu" ou no bucket de "desceu". Isso é conhecido como um problema de <i>classificação</i>. Um modelo que pudesse prever corretamente a direção que o mercado se move seria muito útil.

### O painel da esquerda mostra dois boxplots das mudanças percentuais dos últimos dias no estoque: uma de 648 dias, o qual o mercado aumentou no dia seguinte e uma de 602 dias a qual o mercado diminuiu. Os dois plots parecem praticamente idênticos, sugerindo que não existe estratégia simples utilizando os movimentos de ontem para prever os resultados de hoje. Os outos painéis também indicam pouca associação entre os retornos do passado e do presente. Óbviamente essa falta de padrão era esperada: na presença de fortes correlações entre dias sucessivos, qualquer pessoa poderia adotar uma estratégia simples de trading para gerar lucros do mercado. Entretanto, no capítulo 4, exploramos esses dados utilizando diferentes métodos estatísticos. Curiosamente, existem dicas de algumas tendências fracas em dados que sugerem que, pelo menos no período de 5 anos estudado, é possível prever corretamente a direção do movimento no mercado em aproximadamente 60% do tempo.

### <b>Dados de Expressão Genética</b>

### As duas últimas análises ilustram datasets com variaveis de input e output. Porém, outra importante classe de problemas envolve situações das quais nós somente observamos variáveis de input, sem output correspondente. Por exemplo, numa configuração de mercado, nos queremos ter informações demográficas para um número de clientes (atuais ou em potencial). Nós possivelmente queremos entender que tipos de clientes são similares agrupando-os de acordo com as características observadas. Isso é conhecido como um problema de clustering (agrupamento). Diferentemente dos exemplos anteriores, não estamos tentando prever uma variável de output.

### Por exemplo, em uma análise de células cancerígenas, tentou-se analisar os dados de forma que fosse possível agrupar os 14 tipos de cânceres analisados.
